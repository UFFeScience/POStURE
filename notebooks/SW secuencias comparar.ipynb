```python
import numpy as np
 
def smith_waterman_distance(seq1, seq2, match=3, mismatch=-3, insertion=-2, deletion=-2, normalize=1):
    '''simple and general smith waterman distance for NLP feature extraction'''
    # switch sequences, so that seq1 is the longer sequence to search for seq2
    tsq1=seq1
    tsq2=seq2
    #print(len(seq1),len(seq2))
    if len(seq2)>len(seq1):
        seq1=tsq2
        seq2=tsq1
    # create the distance matrix
    mat = np.zeros((len(seq2) + 1, len(seq1) + 1))
    #print(mat.shape)
    # iterate over the matrix column wise
    for i in range(1, mat.shape[0]):
        # iterate over the matrix row wise
        for j in range(1, mat.shape[1]):
            # set the current matrix element with the maximum of 4 different cases
            mat[i, j] = max(
                # negative values are not allowed
                0,
                # if previous character matches increase the score by match, else decrease it by mismatch
                mat[i - 1, j - 1] + (match if seq1[j - 1] == seq2[i - 1] else mismatch),
                # one character is missing in seq2, so decrease the score by deletion
                mat[i - 1, j] + deletion,
                # one additional character is in seq2, so decrease the scare by insertion
                mat[i, j - 1] + insertion
            )
    # the maximum of mat is now the score, which is returned raw or normalized (with a range of 0-1)
    print(mat)
    print('max',np.max(mat),'tam',len(seq1),'match',match,len(seq1) * match)
    return np.max(mat) / (len(seq1) * match) if normalize else np.max(mat)
    #return np.max(mat)/(len(seq1))
```

```python
from gensim import models, corpora
from operator import itemgetter
dname = "/home/kid/Documentos/analisis2018/lda_models/candidatos45.dict"
mname = "/home/kid/Documentos/analisis2018/lda_models/lda_model_precandi_m60" #model name

#print dname

verbose = True
inverse_topic_word_map = {} #inverse topic hash map
unsaved_docs = 0
inverse_hashmap_word_lookup_length = 100

#loading dictionary and model 
print ("loading dictionary and model")
dictionary = corpora.Dictionary.load(dname)
model = models.LdaModel.load(mname)
```

```python
def create_inverse_hashmap(number_of_topics):
    #Generate the inverse topic hashmap
    for topicid in range(number_of_topics):
        #Create a list of word prob tuples for a topic
        topicid = int(topicid)
        #print(str(topicid))
        topic = model.state.get_lambda()[topicid] #Get words for this topic ***
        #print(topicid,'pollo',topic[:100])
        topic = topic / topic.sum()
        #print('pollo2',topic[:5])
        word_distribution_topic = [(dictionary[id], topic[id]) for id in range(len(dictionary))]
        #Use the tuple to create the hashmap
        #print('pollo3',len(word_distribution_topic))
        #print('pollo4',word_distribution_topic[:5])
        for word, word_probability in word_distribution_topic:
            if not word in inverse_topic_word_map:
                inverse_topic_word_map[word] = [(topicid, word_probability)]
            else:
                inverse_topic_word_map[word].append((topicid, word_probability))

    #print('pollo5',inverse_topic_word_map)
    #Sorting, trimming
    for word in inverse_topic_word_map:
        #Sort the inverse topic hashmap
        inverse_topic_word_map[word] = sorted(inverse_topic_word_map[word] ,key=itemgetter(1))[:inverse_hashmap_word_lookup_length] #probabilidad d palabra a cada tema
    #print(inverse_topic_word_map)
```
