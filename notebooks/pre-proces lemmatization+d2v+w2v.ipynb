```python
import gensim
import os
import re
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim.models.doc2vec import TaggedDocument
from nltk.stem import RSLPStemmer
import spacy
nlp = spacy.load('pt')
import numpy as np
import pandas as pd
```

```python
def load_stopwords( inpath = "stopwords_portu.txt" ):
    stopwords = set()
    with open(inpath) as f:
        lines = f.readlines()
        for l in lines:
            l = l.strip()
            if len(l) > 0:
                stopwords.add(l)
    return stopwords
```

```python
titles=[]
doc_name=[]
doc_weigth=[]
def get_doc_list2(folder_name):
    doc_list = []
    file_list = [folder_name+'/'+name for name in os.listdir(folder_name) if name.endswith('txt')]
    for file in file_list:
        #print(file)
        st = open(file,'r').read()
        doc_list.append(st)
        
    print ('Found %s documents under the dir %s .....'%(len(file_list),folder_name))
    return doc_list

def get_doc_list(folder_name):
    doc_list=[]
    pa='/home/kid/Documentos/analisis2018/dynamic-nmf-master-2018/data/sample_valido/'
    for dirName, subdirList, fileList in os.walk(pa, topdown=False):
        candidato=dirName.split(os.path.sep)[-2][:4]
        tiempo=dirName.split(os.path.sep)[-1][:3]
        ide=candidato+'_'+tiempo
        #print()
        #print('Found directory: %s' % dirName)
        #file_list.append(dirName+'/'+name)
        for file in fileList:
            peso=os.path.getsize(dirName+'/'+file)
            st = open((dirName+'/'+file),'r').read()
            doc_list.append(st)
            #print('->'+ide+'_'+(file[:2]).split('.')[0]+'<-')
            doc_name.append(ide+'_'+(file[:2]).split('.')[0])
    return doc_list

def get_doc(folder_name):
 
    doc_list = get_doc_list(folder_name)
    #print('tamm: '+str(len(doc_list)))
    tokenizer = RegexpTokenizer(r'\w+')
    #nltk.download('stopwords')
    en_stop = load_stopwords("stopwords_portu.txt")
    p_stemmer = PorterStemmer()
 
    text_final = []
 
    texts = []
    for index,i in enumerate(doc_list):
        print(index)
        # for tagged doc
        wordslist = []
        tagslist = []
 
        # clean and tokenize document string
        raw = i.lower()
        tokens = tokenizer.tokenize(raw)
 
        # remove stop words from tokens
        stopped_tokens = [i for i in tokens if not i in en_stop]
 
        # remove numbers
        number_tokens = [re.sub(r'[\d]', ' ', i) for i in stopped_tokens]
        
        text= ' '.join(number_tokens)
        
        lem= nlp(text)
        
        lemmas = [token.lemma_ for token in lem]    
        
        #number_tokens = ' '.join(number_tokens).split()
        
        # stem tokens
        #st = RSLPStemmer()
        #stemmed_tokens = [st.stem(i) for i in number_tokens]
        # remove empty
        #length_tokens = [i for i in stemmed_tokens if len(i) > 1]
        # add tokens to list
        #texts.append(length_tokens)
        #td = TaggedDocument(gensim.utils.to_unicode(str.encode(' '.join(stemmed_tokens))).split(),tags=[str(index)])
        #td = TaggedDocument(gensim.utils.to_unicode(str.encode(' '.join(number_tokens))).split(),tags=[doc_name[index]])
        # for later versions, you may want to use: td = TaggedDocument(gensim.utils.to_unicode(str.encode(' '.join(stemmed_tokens))).split(),[str(index)])
        #print(td)
        text_final.append(' '.join(lemmas))
 
    return text_final
```
